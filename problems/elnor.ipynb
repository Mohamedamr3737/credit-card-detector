```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
import math
# Load the image
def showImage(name, image, cmap=None):
    plt.imshow(image, cmap=cmap)
    plt.axis('off')  # Hide axis
    plt.title(name)
    plt.show()

img = cv2.imread('14 - 2el noor 2ata3.jpg', cv2.IMREAD_GRAYSCALE)
#09 - El spero spathis we23et 3aaaa.jpg
showImage("edges",img)
thresh = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 8)
showImage("",thresh)
edges = cv2.Canny(thresh, 50, 150)  # Adjust thresholds as needed
showImage("edges",edges)
```
